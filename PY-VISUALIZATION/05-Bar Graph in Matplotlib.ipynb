## Bar graph arr used to compare numeric values for specific categories
### Problem: Plot top 10 Countries by population 

```python
import pandas as pd
from matplotlib import pyplot as plt
data = pd.read_csv('datasets/countries.csv')
data.head()
```

```python
top10_2007 = data[data.year == 2007].sort_values('population', ascending=False).head(10)
top10_2007
```

### Lets plot on bar graph

```python
x= range(10) # [0 .. 9] for x axis

# Divide population by 10**6 to show plot in millions
plt.bar(x, top10_2007.population / (10 ** 6))

# Add labels
plt.xticks(x, top10_2007.country, rotation=45)
plt.title("Top 10 countries with highest population")
plt.ylabel("Population (in millions)")
plt.xlabel("Country")
plt.show()
```

## Lets plot population along with GDP for these top 10 countries

```python
import numpy as np
x = np.arange(10)
# We have to create graph with multiple axes
fig, ax = plt.subplots()

# This is width of each bar
width = 0.3

plt.xticks(x, top10_2007.country, rotation=45)
plt.title("Top 10 countries with highest population and their GDP")
plt.xlabel("Country")

# First Graph with Population
population = ax.bar(x, top10_2007.population / (10 ** 6), width)
plt.ylabel("Population (in millions)")

# Second Graph with GDP. Twinx reuse the same X axis but create new Y axis
ax2 = ax.twinx()
gdp = ax2.bar(x + width, (top10_2007.population * top10_2007.gdpPerCapita) / (10 ** 9), width, color = 'orange')
plt.ylabel("GDP (in billions)")

# Add legend. We will use the population and gdp object returned by ax.bar and ax2.bar for setting legend
plt.legend([population, gdp], ["Population", "GDP"])
plt.show()
```
