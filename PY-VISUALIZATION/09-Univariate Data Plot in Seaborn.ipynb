# Univariate data
This is data with single variable. E.g. List of weights of adults etc.  
We can use following graphs to visualize data distribution and trends in data.  
 - Historam
 - KDE Plot
 - Rug Plot

# KDE - Kernel Density Estimation
![KDE](images/kde.png)

## Lets plot Histogram and KDE plot on alcohol content of wines

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('datasets/winequality-white.csv', 
                   names=['fixed_acidity', 'volatile_acidity', 'citric_acidity', 'residual_sugar',
                          'chlorides', 'free_sulphur_dioxide', 'total_sulphur_dioxide', 'density',
                          'ph','sulphates','alcohol', 'quality'], 
                   skiprows=1,
                   sep=r'\s*;\s*', engine='python')
data.head()
```

```python
sns.displot(data=data.alcohol, kde=True, height=5, aspect=3)
plt.show()
```

#### Now we can also see individual data points on this plot using rug plot

```python
sns.displot(data=data.alcohol, kde=True, rug=True, height=5, aspect=3)
plt.show()
```

Here we are seeing small lines on at bottom which are individual data points

#### We can only see rug / individual data points graph using rugplot

```python
sns.rugplot(data=data.alcohol, height=0.5)
plt.show()
```

#### Lets say we want to see KDE and rug plots but not histogram

```python
sns.set(color_codes=True)
sns.displot(data=data.alcohol, rug=True, kind='kde', height=5, aspect=2, fill=True, color='r')
plt.show()
```
