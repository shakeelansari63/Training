## Import modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
import scipy
```

### Load Dataset

```python
df = pd.read_csv('facebook_data.csv')
df.head()
```

### Clean

### Line Chart

```python
data = df[['tenure', 'likes_received']].groupby('tenure').mean().reset_index()
sns.lineplot(data=data, x='tenure', y='likes_received')
plt.show()
```

### Bar Chart

```python
data = df[['gender', 'likes_received']].groupby('gender').sum().reset_index()
sns.barplot(data=data, x='gender', y='likes_received')
plt.show()
```

### Histogram

```python
sns.displot(data=df['likes_received'], bins=100, kde=True)
plt.xlim(0,50000)
plt.ylim(0, 5000)
plt.show()
```

### Pie Chart

```python
data = df[['gender', 'likes_received']].groupby('gender').sum().reset_index()
plt.pie(data['likes_received'], labels=data['gender'])
plt.show()
```

### Bubble Chart

```python
data = df[['age', 'likes_received', 'friend_count']]
sns.scatterplot(data=data, x='age', y='likes_received', hue='friend_count', size='friend_count')
plt.show()
```
