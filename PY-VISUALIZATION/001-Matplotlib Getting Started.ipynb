## Matplotlib is python library to create plots

```python
# Import Pyplot from matplotlib
from matplotlib import pyplot as plt
```

### Now we can create some simple plots using plt object

```python
plt.plot([1,2,3], [1,4,9]) # plot data
plt.xlabel("X") # X Axis Label
plt.ylabel("Y=X^2") # Y Axis label
plt.show() # Shows the plot
```

### we can plot multiple lines 

```python
plt.plot([1,2,3], [1,4,9]) # plot data series 1
plt.plot([1,2,3], [1,8,27]) # plot data series 2
plt.legend(['Data: X^2', 'Data: X^3'])  # Legend
plt.xlabel('X') # X Label
plt.show()
```

## Matplotlib can use pandas to plot graphs

```python
# Import Pandas
import pandas as pd

# Create sample Dataframe
data = {
    "year": [2001, 2005, 2009, 2013, 2017, 2021],
    "attendees": [123, 173, 120, 145, 178, 189],
    "avg_age": [23, 25, 32, 24, 28, 30]
}
df = pd.DataFrame(data)
df
```

```python
# Plot Df with matplotlib
plt.plot(df['year'], df['attendees']) # Plot Attendees
plt.plot(df['year'], df['avg_age']) # Plot Average age
plt.legend(['Attendees', 'Average Age'])
plt.show()
```
