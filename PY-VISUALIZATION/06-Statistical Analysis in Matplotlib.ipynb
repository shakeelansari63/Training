# Data Distribution Analysis with Histogram
Analysing data distribution is very common technique to understand data in Statistcal Analysis.  
And Histogram is a most basic graph which is used to analyse the data distribution. By plotting Histogram, we can observe how the data is distributed.    
   
![Data Distribution](images/histogram-data-distribution.png)

## Anatomy of a histogram
![Anatomy of histogram](images/anatomy-of-histogram.png)

### Why finding optimal number of bins is necessary for histogram?
![Optimise bin numbers](images/histogram-bins.png)

### Lets generate some random numbers using numpy.randon.default_rng for visualization

```python
import numpy as np
from matplotlib import pyplot as plt

# Default seed for random number generator to generate same numbers always
rng = np.random.default_rng(seed = 112)

# Generate a list of random numbers from Normal Distribution
# We take 500 numbers where mean location is 150 and data is distributed about +-10 from mean 
# (i.e. standard deviation is 10)
data = rng.normal(loc=150, scale=10, size=500)

# Plot Histogram
plt.hist(data)
plt.show()
```

### Lets check some options available in Histogram

#### Bins
This option allows to define the number of bins

```python
plt.figure(figsize=(12, 6))
plt.subplot(211)
# Default Plot
plt.hist(data)
plt.title("Original Histogram")

# Plot with bins in number
plt.subplot(223)
plt.hist(data, bins=15)
plt.title("15 Bins")

# Plot with bins identified by SQRT funtion
plt.subplot(224)
plt.hist(data, bins='sqrt')
plt.title("SQRT Bins")

# Show PLot
plt.tight_layout()
plt.show()
```

#### Density
We can define histogram to show Probability Density (Probability that a value would fall in this bin) instead of actual counts

```python
plt.figure(figsize=(10, 3))
plt.subplot(121)
# Default Plot
plt.hist(data)
plt.title("Original Histogram")

# Plot with density
plt.subplot(122)
plt.hist(data, density=True)
plt.title("Probability Density")

# Show PLot
plt.tight_layout()
plt.show()
```

#### Cumulative
Cumulative option shows the histogram for cumulative data. Then the data is no more for specific bins, but for bins below current bin

```python
plt.figure(figsize=(12, 6))
plt.subplot(211)
# Default Plot
plt.hist(data)
plt.title("Original Histogram")

# Plot with cumulative data
plt.subplot(223)
plt.hist(data, cumulative=True)
plt.title("Cumulative Histogram")

# Plot with cumulative density data
plt.subplot(224)
plt.hist(data, density=True, cumulative=True)
plt.title("Cumulative probability density")

# Show PLot
plt.tight_layout()
plt.show()
```

As you can see, since we have 500 data points, the cumulative histogram on y axis goes to 500  
And Since Probability has max value as 1, cumulative probability density histogram has max value of Y as 1.0

#### We can also adjust some visual parameters for styling histogram

```python
plt.figure(figsize=(12, 6))
plt.subplot(221)
# Default Plot
plt.hist(data)
plt.title("Original Histogram")

# Plot with orange color bars
plt.subplot(222)
plt.hist(data, color="orange")
plt.title("Colored Historam")

# Plot with just outline as steps
plt.subplot(223)
plt.hist(data, histtype='step')
plt.title("Stepped Historam")

# Plot with gaps in bars
plt.subplot(224)
plt.hist(data, rwidth=0.8)
plt.title("Bar width for histogram")

# Show PLot
plt.tight_layout()
plt.show()
```

# Data Analysis with Box Plot

## Anatomy of a Boxplot
![Anatomy of Boxplot](images/anatomy-of-boxplot.png)

### Let's see what these values mean by seeing the data aggregated data in our data points

```python
import pandas as pd
pd.Series(data).describe()
```

```python
plt.boxplot(data, showmeans=True)
plt.show()
```

#### Here as we can see, 
##### Mean is indicated with central triangle
##### Median (50%) is yellow line
##### 25% is lower limit of box
##### 75% is upper limit of Box
##### Whisker is calculated by default as 1.5xIQR
![Whisker Calculation](images/boxplot-whisker.png)
##### And circles are outliers which are outside whisker

## The above is for Normal distribution, If we plot Exponential Distribution, the box plot and histogram would look as follow

```python
# Default seed for random number generator to generate same numbers always
rng = np.random.default_rng(seed = 112)

# Generate a list of random numbers from Exponential Distribution
data2 = rng.exponential(scale=1, size=500)
```

```python
plt.figure(figsize=(10,3))

# Histogram
plt.subplot(121)
plt.hist(data2)
plt.title("Histogram of Exponeential Distribution")

# Box Plot
plt.subplot(122)
plt.boxplot(data2, showmeans=True)
plt.title("Boxplot of Exponeential Distribution")

# Show
plt.tight_layout()
plt.show()
```

## Distinguish Skewed distribution
![Skewed Distribution](images/plots-skewed-distribution.png)

## With Box plot, we can show a notch near Median line which can be useful when comparing multiple box plots in single plot

```python
# Default seed for random number generator to generate same numbers always
rng = np.random.default_rng(seed = 112)

# generate random integers
data3 = rng.integers(low=1, high=10, size=(50, 3))
```

```python
plt.figure(figsize=(10, 3))
plt.boxplot(data3, tick_labels=['A', 'B', 'C'], notch=True)
plt.show()
```

### We can also show mean line instead of mean marker

```python
plt.figure(figsize=(10, 3))
plt.boxplot(data3, tick_labels=['A', 'B', 'C'], showmeans=True, meanline=True)
plt.show()
```

# Violin Plot
Violin Plot is similar to Box plot as it plots summary of data, but it also provides some extra infomation

## Anatomy of Violin plot
![Anatomy of Violinplot](images/anatomy-of-violinplot.png)

## Lets plot violin plot on our normal distributed data

```python
plt.violinplot(data, showmeans=True, showmedians=True)
plt.show()
```

### If we plot same on Exponentially distributed data, we will see something like 

```python
plt.violinplot(data2, showmeans=True, showmedians=True)
plt.show()
```

## Lets plot Box and violin plots side by side with histogram and see

```python
plt.figure(figsize=(12, 6))

# Histogram
plt.subplot(221)
plt.hist(data)

# Box Plot
plt.subplot(223)
plt.boxplot(data, showmeans=True, meanline=True, vert=False)

# Violin Plot
plt.subplot(122)
plt.violinplot(data, showmeans=True, showmedians=True)

# Show
plt.tight_layout()
plt.show()
```

As we can see histogram is good for visualizing data distribution, Box plot is good for visualizing summary.   
But Violin plot show both in same plot

## Lets plot same on Exponential data and check

```python
plt.figure(figsize=(12, 6))

# Histogram
plt.subplot(221)
plt.hist(data2)

# Box Plot
plt.subplot(223)
plt.boxplot(data2, showmeans=True, meanline=True, vert=False)

# Violin Plot
plt.subplot(122)
plt.violinplot(data2, showmeans=True, showmedians=True)

# Show
plt.tight_layout()
plt.show()
```

Now Lets create a bimodal / multimodal data and see these plots. And the information will be clear there

```python
# Default seed for random number generator to generate same numbers always
rng = np.random.default_rng(seed = 112)

# We chose 2 data series with different mean and standard deviation
d1 = rng.normal(loc=150, scale=10, size=500)
d2 = rng.normal(loc=230, scale=5, size=500)

data4 = np.concatenate([d1, d2])
```

```python
plt.figure(figsize=(12, 6))

# Histogram
plt.subplot(221)
plt.hist(data4, bins=25)

# Box Plot
plt.subplot(223)
plt.boxplot(data4, showmeans=True, meanline=True, vert=False)

# Violin Plot
plt.subplot(122)
plt.violinplot(data4, showmeans=True, showmedians=True)

# Show
plt.tight_layout()
plt.show()
```

#### So we can see that the Histogram can tell us that data is bimodal, but it does not tell mean and medians
#### And Box shows mean, median and other summaries, but there is no way to find out modality.
#### And this is where violin comes in, it shows both summaries with data modality

## We can also display above graph using subplot_mosaic
### It gives us better control on where we put what and also some sizing
### 
### For example, lets say we need following graph, now the box plt is smaller than histogram. So we have devided it into 3 rows
![Mosaic Subplots](images/mosaic.png)

```python
fig = plt.figure(tight_layout=True)

# Create mosaic
axes = fig.subplot_mosaic([['box', 'violin'],
                           ['hist', 'violin'],
                           ['hist', 'violin']])

# Notice we have given same names in 2nd and 3rd row to reserve space for same graph
# Plot Box
axes['box'].boxplot(data4, vert=False)
axes['box'].set_title('Box Plot')

# Plot Histogram
axes['hist'].hist(data4, bins=25)
axes['hist'].set_title('Histogram Plot')

# Plot Violin
axes['violin'].violinplot(data4, showmeans=True, showmedians=True)
axes['violin'].set_title('Violin Plot')

# Show plots
plt.suptitle("Multiple Plots for bimodal data")
plt.show()
```
