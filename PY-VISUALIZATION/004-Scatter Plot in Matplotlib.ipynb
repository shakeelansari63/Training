## Scatter plot provides relation between multiple variables
### Problem: Compare gdpPerCapita and Life Expectancy 

```python
import pandas as pd
from matplotlib import pyplot as plt
data = pd.read_csv('countries.csv')
data.head()
```

## Let us compare gdp per capita and life expectancy in year 2007 which is latest year

```python
plt.scatter(data[data.year == 2007].gdpPerCapita, data[data.year == 2007].lifeExpectancy)
plt.show()
```

### As we can see, the Life expectancy grows exponentially in initial growth of gdpPerCapita, but that growth slows down.

#### we can use log scale to see the linear data

```python
import numpy as np
## we will use np.log10 to take log base 1o of gdppercapita
plt.scatter(np.log10(data[data.year == 2007].gdpPerCapita), data[data.year == 2007].lifeExpectancy, 5)
plt.show()
```

And this shows lines correlation. 

## Lets show all years data


```python
years = sorted(list(set(data.year)))
```

```python
import math
columns = 3 # Number of columns in subplot
rows = math.ceil(len(years) / columns) # Number of rows needed in subplot
for i, year in enumerate(years):
    plt.subplot(rows, columns, i+1)
    plt.subplots_adjust(wspace=1, hspace=1.4)
    plt.scatter(data[data.year == year].gdpPerCapita, data[data.year == 2007].lifeExpectancy, 2)
    plt.ylim(25, 90) # Limit y range
    plt.title(f'Year: {year}')
plt.suptitle("Correlation between GDPPerCapita and Life Expectancy for all years")
plt.show()
```

#### We can also plot them with Log scale

```python
import math
columns = 3 # Number of columns in subplot
rows = math.ceil(len(years) / columns) # Number of rows needed in subplot
for i, year in enumerate(years):
    plt.subplot(rows, columns, i+1)
    plt.subplots_adjust(wspace=1, hspace=1.4)
    plt.scatter(np.log10(data[data.year == year].gdpPerCapita), data[data.year == 2007].lifeExpectancy, 2)
    plt.ylim(25, 90) # Limit y range
    plt.title(f'Year: {year}')
plt.suptitle("Correlation between GDPPerCapita and Life Expectancy for all years")
plt.show()
```
