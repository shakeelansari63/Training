## Here we will plot graph directly from Pandas Dataframe
### Lets read an xl file with some data first

```python
import pandas as pd
from matplotlib import pyplot as plt

df = pd.read_excel('datasets/lures.xlsx', sheet_name='LURES')
df.head()
```

### Lets plot the sales on scatter plot

```python
df.plot(kind='scatter', x = 'QUANTITY', y = 'SALES', title = "Sales vs Quantity", marker = "^", color = "orange")
```

### We can create more charts from this data

```python
lures_sales_by_region = df.groupby(['REGION'], as_index=False)['SALES'].sum()
lures_sales_by_region
```

```python
lures_sales_by_region.plot(kind = 'bar', x = 'REGION', y = 'SALES')
```

### Or we can plot time series data as line chart

```python
df.plot(kind = 'line', x = 'DATE', y = 'SALES')
```

### But the data is very much scattered, we can aggregate the data to show it better

```python
lures_by_day = df.groupby(['DATE'], as_index=False)['SALES'].sum()
lures_by_day.head()
```

```python
lures_by_day.plot(kind = 'line', x = 'DATE', y = 'SALES')
```
