## Matplotlib: Basic Plotting using pyplot

```python
## import Numpy
import numpy as np

## Import Matplotlib 
import matplotlib.pyplot as plt

## Generate Numpy x-coordinates with range having step size 1
x = np.arange(-100, 101, 1)

## f(x) is y with function of x
y = x **2 + 2*x + 4

## Create Plot
plt.plot(x, y)
plt.show()
```

## Plotting multiple functions in same plot

```python
## import Numpy
import numpy as np

## Import Matplotlib 
import matplotlib.pyplot as plt

## Generate Numpy x-coordinates with range having step size 1
x = np.arange(-100, 101, 1)

## f(x) is y with function of x
y1 = x **2 + 2*x + 4
y2 = 2000 * np.sin(x)
y3 = 2000 * np.log(abs(x))

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

plt.show()
```

## Styling plot
Plot can be styled by 3rd parameter which tells line colour and pattern  
e.g.  
`r-` $\to$ Red colour complete line  
`r--` $\to$ Red colour dashed line   
`r-.` $\to$ Red colour line with dash and dot  
`r.` $\to$ Red colour dotted line  
`ro` $\to$ Red colour stroked dotted line  

```python
## import Numpy
import numpy as np

## Import Matplotlib 
import matplotlib.pyplot as plt

## Generate Numpy x-coordinates with range having step size 1
x = np.arange(-20, 21, 0.5)

## f(x) is y with function of x
y1 = x **2 + 4
y2 = 20 * np.sin(x)
y3 = 100 * np.log(abs(x))

plt.plot(x, y1, 'r--')
plt.plot(x, y2, 'g-.')
plt.plot(x, y3, 'b.')

plt.show()
```

## Subplot
Subplot is having multiple plots on same window.

```python
import numpy as np
import matplotlib.pyplot as plt

## X Co-ordinates
x = np.arange(0, 101, 0.1)

## Y is f(x)
y1 = 4*(x**2) + 4
y2 = np.sin(x)


## Create Subplot
## Here subplot take 3 digit number
## 1st Digit (100th place) - Number of subplot rows
## 2nd Digit ( 10th Place) - Number of subplot columns
## 3rd Digit ( Unit Place) - Number for this subplot
sp1 = plt.subplot(211) ## There are 2 rows and 1 columns of subplot and this is 1st subplot
sp2 = plt.subplot(212) ## There are 2 rows and 1 columns of subplot and this is 2nd subplot

## Plot the subplots
sp1.plot(x, y1)
sp2.plot(x, y2)

## Tight layout for correct spacings
## Without this the x/y co-ordinate values may overflow on plot beside
plt.tight_layout()

## Show the pyplot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

## X Co-ordinates
x = np.arange(-10, 11, 0.1)

## Y is f(x)
y1 = 4*(x**2) + 4
y2 = np.sin(x)
y3 = np.log(x)
y4 = x**3

## Create Subplot
## Here subplot take 3 digit number
## 1st Digit (100th place) - Number of subplot rows
## 2nd Digit ( 10th Place) - Number of subplot columns
## 3rd Digit ( Unit Place) - Number for this subplot
sp1 = plt.subplot(221) ## There are 2 rows and 2 columns of subplot and this is 1st subplot
sp2 = plt.subplot(222) ## There are 2 rows and 2 columns of subplot and this is 2nd subplot
sp3 = plt.subplot(223) ## There are 2 rows and 2 columns of subplot and this is 3rd subplot
sp4 = plt.subplot(224) ## There are 2 rows and 2 columns of subplot and this is 4th subplot

## Plot the subplots
sp1.plot(x, y1, 'r-')
sp2.plot(x, y2, 'g-.')
sp3.plot(x, y3, 'b.')
sp4.plot(x, y4, 'y--')

## Tight layout for correct spacings
## Without this the x/y co-ordinate values may overflow on plot beside
plt.tight_layout()

## Show the pyplot
plt.show()
```

## Multiple Plot Windows
Subplots are multiple plots on same window. But if we need different windows for each different plot, we have to define figures.

```python
import numpy as np
import matplotlib.pyplot as plt

## X Co-ordinates
x = np.arange(-10, 11, 0.1)

## Y is f(x)
y1 = 4*(x**2) + 4
y2 = np.sin(x)
y3 = np.log(x)
y4 = x**3

## Figure 1 i.e. 1st window
plt.figure(1)

## This Figure / window has 2 subplots
sp1 = plt.subplot(211) ## There are 2 rows and 1 columns of subplot and this is 1st subplot
sp2 = plt.subplot(212) ## There are 2 rows and 1 columns of subplot and this is 2nd subplot

## Plot the subplots
sp1.plot(x, y1, 'r-')
sp2.plot(x, y2, 'g-.')

## Tight layout for subplots
plt.tight_layout()

## Figure 2
plt.figure(2)

## This Figure has only 1 plot
plt.plot(x, y3, 'b.')

## Figure 3
plt.figure(3)

## This Figure has only 1 plot
plt.plot(x, y4, 'y--')

## Show the pyplot
plt.show()
```

## Matplotlib Styling
matplotlib.style has set of predefined styles which can be applied to plots easily.  
  
Visit [Matplotlib Style Sheet Reference](https://matplotlib.org/3.1.0/gallery/style_sheets/style_sheets_reference.html) for details on all available plots

```python
import numpy as np
import matplotlib.pyplot as plt

## Styling libraray
from matplotlib import style as sy

## Define function
x = np.arange(-1, 11, 0.1)
y = np.sin(x)

## Use predefined GGPLOT style
sy.use('ggplot')

plt.plot(x, y)
plt.show()
```

### Custom Styling
You can do custom styling instead of using the predefined styles

```python
import numpy as np
import matplotlib.pyplot as plt

## Styling libraray
from matplotlib import style as sy

## Define function
x = np.arange(-10, 10.1, 0.1)
y = np.sin(x)

## Use predefined Default style which is most basic style
sy.use('default')

## Default style doen not have grid, to show grid
plt.grid(True)

## To display Title for Graph
plt.title("Sine Function")

## To display SupTitle for Figure
plt.suptitle("Freak Graph")

## Labeling the Axes

## X Label
plt.xlabel("Time")
## Y Label
plt.ylabel("Amplitude")

plt.plot(x, y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

## Styling libraray
from matplotlib import style as sy

## Define function
x = np.arange(-10, 10.1, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

## Use predefined Default style which is most basic style
sy.use('default')

## To display Title for Graph
plt.title("Sine Function")

## Labeling the Axes

## X Label
plt.xlabel("Time")
## Y Label
plt.ylabel("Amplitude")

## Display Legends
## In order to display legends, the function plots need to be labeled

plt.plot(x, y1, 'r--', label = "Sine")
plt.plot(x, y2, 'y-', label = "Cosine")

## Display Legend
plt.legend(loc = 'upper right')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

## Build data for bar chart
## Here we have skill level of 4 individuals in different skillsets on scale of 100
skill = {
    'python': (27, 44, 69, 81),
    'java': (59, 78, 45, 36),
    'networking': (72, 63, 54, 67),
    'machine_learning': (22, 37, 74, 77)
}

## Name of the individuals
people = ['Bob', 'John', 'Martha', 'Liz']

## Now we need X values which will be used to plot on X axis. Here it is dummy key.
## Since there are 4 individuals, we take 4 item at step of 2
idx = np.arange(0, 8, 2)

## Lets Plot the Bar chart
## We provide Width so we can control the position of subsequent plots
## We also define label for legend
plt.bar(idx, skill['python'], width = 0.2, label = "Python")
plt.bar(idx+0.2, skill['java'], width = 0.2, label = "Java") ## X index right moved by 0.2
plt.bar(idx+0.4, skill['networking'], width = 0.2, label = "Networking") ## X index right moved by 0.4
plt.bar(idx+0.6, skill['machine_learning'], width = 0.2, label = "Machine Learning") ## X index right moved by 0.6

## Display Y Labels
plt.ylabel("Skill level")

## Lets not display X label, as that does not make sence. 
## The values are not continous, they are categorical. 
## So we use Xtick to put categories
plt.xticks(idx+0.3, people)

## Since all the values are below 90, the bar graph will display y axis only till 90.
## Lets set limit manually.
plt.ylim(0, 120)

## Display the Legend
plt.legend()

plt.show()
```

## Pie Charts

```python
import matplotlib.pyplot as plt

## Data for basic Pie chart
data = [20, 56, 71, 92]
labels = ['Grade A', 'Grade D', 'Grade C', 'Grade B']

plt.pie(data, labels = labels)
plt.show()
```

```python
import matplotlib.pyplot as plt

## Data for basic Pie chart
num_of_students_in_grades = [20, 71, 92, 56]
grades = ['Grade A', 'Grade C', 'Grade B', 'Grade D']

## We can explode out the grade A students by explode list. 
## This has same data as actual actual list and part to explode having non zero value
explode = [0.1, 0.01, 0.01, 0.01]

## We can Show percentage automatically using autopct
## Can state starting angle. By default starting angle is 0
## If we want teh grapth to come Clockwise, then we disable counter clock
plt.pie(num_of_students_in_grades
        , labels = grades
        , autopct = '%.2f%%'
        , explode = explode
        , startangle=90
        , counterclock = False)

## Display Title
plt.title("Students Grades")

plt.show()
```

## Histogram

```python
import numpy as np
import matplotlib.pyplot as plt

## Lets have random data for visualization
data = np.random.randn(10000)

## Draw Basic histogram
plt.hist(data, 100)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

## Here is statistics of students heights
mu, sigma = 177, 8 # mu is mean height and sigma is standard deviation from mean
heights = mu + sigma * np.random.randn(10000)

## Draw Basic histogram
## Y Axis show the actual values. It can be converted to percentage with density
plt.hist(heights, 100
         , facecolor = 'green'
         , density = True)

## Labels
plt.xlabel('Students Heights')
plt.ylabel('% age')

## Show grid
plt.grid(True)

## Display Mean and Standard Deviation
plt.text(150, 0.04, f" = {mu}\n = {sigma}")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

## 3D Plotting Library
from mpl_toolkits import mplot3d

## Define 3D co-ordinate system
proj = plt.axes(projection = '3d')

## We need 3 co-ordinate system values for plotting in 3d.
## We can take 1 co-ordinate as inp and other calculated
x = np.arange(0.01, 50, 0.01)
y = np.sin(x)
z = np.cos(x)

## Create plot
## Here we use axes pjection to plot instead of pyplot
proj.plot3D(x, y, z)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

## 3D Plotting Library
from mpl_toolkits import mplot3d

## Define 3D co-ordinate system
proj = plt.axes(projection = '3d')

## We need 3 co-ordinate system values for plotting in 3d.
## We can take 2 co-ordinates as inp and third calculated
x = np.arange(0.1, 50, 0.1)
y = np.arange(0.1, 50, 0.1)

## In order to build a surface instead of line, we can mesh grids in numpy and make a matrix
X, Y = np.meshgrid(x, y)

Z1 = np.sqrt(X **2 + Y**2) ## This will make a plane as z = sqrt(x^2 + y^2)
Z2 = np.sin(Z1)

## Create plot
## Here we use axes pjection to plot instead of pyplot
proj.plot_surface(X, Y, Z1)
proj.plot_surface(X, Y, Z2)

plt.show()
```
