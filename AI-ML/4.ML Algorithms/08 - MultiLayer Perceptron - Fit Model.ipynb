## Multilayer Perceptron: Fit and evaluate a model

Using the Titanic dataset from [this](https://www.kaggle.com/c/titanic/overview) Kaggle competition.

In this section, we will fit and evaluate a simple Multilayer Perceptron model.

### Read in Data

```python
import pandas as pd
from sklearn.model_selection import GridSearchCV
from sklearn.neural_network import MLPClassifier
import warnings
warnings.filterwarnings('ignore', category=FutureWarning)
warnings.filterwarnings('ignore', category=DeprecationWarning)

tr_features = pd.read_csv('tmp/train_features.csv')
tr_labels = pd.read_csv('tmp/train_labels.csv')
```

### Hyperparameter tuning

![hidden layer](img/hidden_layers.png)

```python
def print_results(results):
    print('BEST PARAMS: {}\n'.format(results.best_params_))

    means = results.cv_results_['mean_test_score']
    stds = results.cv_results_['std_test_score']
    for mean, std, params in zip(means, stds, results.cv_results_['params']):
        print('{} (+/-{}) for {}'.format(round(mean, 3), round(std * 2, 3), params))
```

```python
## Create MLP Model
mlp = MLPClassifier()

## Define Hyper Parameters
parameters = {
    'hidden_layer_sizes': [(10,), (50,), (100,)], # 1st element of tuple is nodes and 2nd is layer. Empty layer mean 1
    'activation': ['relu', 'tanh', 'logistic'], # Logistic is same as Sigmoid
    'learning_rate': ['constant', 'invscaling', 'adaptive']
}

## Create 5 Fold Grid Search Cross Validation
cv = GridSearchCV(mlp, parameters, cv = 5)

## Fit training data
cv.fit(tr_features, tr_labels.values.ravel())

## Print results
print_results(cv)
```

```python
## Lets grap the best estimator
print(cv.best_estimator_)
```

### Write out pickled model

```python
import joblib

joblib.dump(cv.best_estimator_, 'joblib/MLP_Model.pkl')
```
